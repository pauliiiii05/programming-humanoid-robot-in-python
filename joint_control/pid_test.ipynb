````python
%matplotlib notebook
from matplotlib import pylab as plt
from ipywidgets import interact
from IPython import display
from collections import deque
````

## start simspark
Start SimSpark simulator as described in [last lecture](../introduction/README.md)

## run PIDAgent
Run all code below, there is plot show current joint angle and desired joint angle.

You can set target value and tune the PID parameters to get better results. 


````python
from pid import PIDAgent
agent = PIDAgent()
````

````python
@interact(kp=(0, 100, 1), ki=(-1, 1, 0.1), kd=(-1, 1, 0.1), delay=(0, 5))
def set_pid_parameters(kp, ki, kd, delay=0):
    global agent
    agent.joint_controller.Kp = kp
    agent.joint_controller.Ki = ki
    agent.joint_controller.Kd = kd
    agent.joint_controller.set_delay(delay)

joint_name = 'HeadYaw'
sensor = deque(maxlen=100)
target = deque(maxlen=100)

@interact(name=agent.joint_names)
def set_joint_name(name):
    global joint_name
    if joint_name != name:
        joint_name = name
        sensor.clear()
        target.clear()

@interact(value=(-1, 1, 0.1))
def set_joint_target(value):
    agent.target_joints[joint_name] = value
````

````python
# inject plotting input agent's loop
cycle = 0
orig_sense_think_act = agent.sense_think_act

def sense_think_act():
    global cycle, joint_name, sensor, target
    cycle += 1
    
    # 1. Run the original agent logic
    orig_sense_think_act()
    
    # 2. JUST COLLECT DATA. NO PLOTTING.
    # We add a try/except in case the joint isn't in the perception yet
    try:
        sensor.append(agent.perception.joint[joint_name])
        target.append(agent.target_joints[joint_name])
    except KeyError:
        pass # Ignore if the joint_name isn't available yet

# 3. Inject the new data-collection function
agent.sense_think_act = sense_think_act

# 4. Start the agent thread. This will return immediately.
agent.start()
print("Agent thread started. Run the next cell to start plotting.")
````

````python
import time

# 1. Create the figure and axes ONCE
fig, ax = plt.subplots()

# 2. This loop runs in the main notebook thread
try:
    while True:
        # Wait until the agent thread has collected some data
        if not sensor: 
            time.sleep(0.1)
            continue
            
        # 3. Plot the data that the other thread collected
        ax.cla()  # Clear the axes
        ax.plot(target, '-r', label='Target')
        ax.plot(sensor, '-b', label='Sensor')
        ax.legend()
        ax.set_title(f'Joint: {joint_name} (Cycle: {cycle})')
        
        # 4. Update the notebook display
        display.clear_output(wait=True)
        display.display(fig)
        
        # 5. Pause briefly to create a smooth animation
        time.sleep(0.1) 

except KeyboardInterrupt:
    # 6. When you stop this cell, it stops the agent
    agent.stop()
    print("Plotting stopped and agent shut down.")
````
